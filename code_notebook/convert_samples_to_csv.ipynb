```python
import os
import pandas as pd
```

```python
# Notebook nằm trong Project/code_notebook/
# Samples nằm trong Project/data/samples/
samples_dir = os.path.join(os.path.dirname(os.getcwd()), 'data', 'samples')
csv_dir = os.path.join(os.path.dirname(os.getcwd()), 'data', 'samples_csv')
os.makedirs(csv_dir, exist_ok=True)
print(f'Input  : {samples_dir}')
print(f'Output : {csv_dir}')
```

```python
# Các kích cỡ cần convert (50 → 500)
target_sizes = [50, 100, 150, 200, 250, 300, 350, 400, 450, 500]

for n_users in target_sizes:
    fname_parquet = f'play_count_sample_{n_users}users.snappy.parquet'
    fpath_parquet = os.path.join(samples_dir, fname_parquet)

    if not os.path.exists(fpath_parquet):
        print(f'[SKIP] File không tồn tại: {fname_parquet}')
        continue

    df = pd.read_parquet(fpath_parquet)

    fname_csv = f'play_count_sample_{n_users}users.csv'
    fpath_csv = os.path.join(csv_dir, fname_csv)
    df.to_csv(fpath_csv, sep='|', index=False)

    print(f'[OK] {fname_parquet} → {fname_csv}  |  shape={df.shape}')
```

```python
# Kiểm tra nhanh file 100 users
check_path = os.path.join(csv_dir, 'play_count_sample_100users.csv')
check = pd.read_csv(check_path, sep='|', nrows=5)
print(check.to_string(index=False))
```
